### 3.
제품의 품질을 조사하여, Cabbage 결함과 결함이 발생한 기간을 조사한 Table입니다.  
범주형 데이터에 대해, counts, percents, cumulative counts, cumulative percents 계산 바랍니다.   
- (Data set : EXH_QC1.csv)

```python
data=pd.read_csv("../data/exh_QC1.csv", engine='python')
data.head()
```

```python
df=data[['Flaws', 'Period']]
df.head()
```

```python
counts=data['Flaws'].value_counts()
percents=counts/sum(counts)*100
cumcnts=np.cumsum(counts)
cumpcts=np.cumsum(percents)
```

```python
print("[data['Flaws']]")
display(data['Flaws'].head())

print('\n[count]')
display(counts)

print('\n[cumcnt]')
display(cumcnts)

print('\n[percent]')
display(percents)

print('\n[cumpct]')
display(cumpcts)
```

```python
# DataFrame으로 취합
count_data=pd.DataFrame({'Count':counts, 'CumCnt':cumcnts, 'Percent':percents, 'CumPct':cumpcts})
# Column의 name(좌측상단 이름) 생성
count_data.columns.name='Flaws'
# 결과 확인
count_data
```

```python
df=data['Period']
counts=df.value_counts()
percents=counts/sum(counts)*100
cumcnts=np.cumsum(counts)
cumpnts=np.cumsum(percents)
```

```python

print("[data['Period']]")
display(data['Period'].head())

print('\n[count]')
display(counts)

print('\n[cumcnt]')
display(cumcnts)

print('\n[percent]')
display(percents)

print('\n[cumpnt]')
display(cumpnts)
```

```python
# DataFrame으로 취합
count_data=pd.DataFrame({'Count':counts, 'CumCnt':cumcnts, 'Percent':percents, 'CumPnt':cumpnts})
# Column의 name(좌측상단 이름) 생성
count_data.columns.name='Period'
# 결과 확인
count_data
```
