### 1.
정규분포에서 추출한 10기이 무작위 데이터로 slice, loc, iloc, query 등을 이용하여 데이터를 선택하라.  
데이터: 정규분포에서 추출한 10개의 무작위 데이터.
- (변수: 2개, 자료 수: 10개, random seed=1234)

```python
# 샘플 데이터 통이를 위해 seed 고정: 1234
np.random.seed(1234)

# 랜덤 함수를 이용해 df데이터 생성
df=pd.DataFrame(randn(10,2), columns=['A', 'B'])
df
```

```python
# 열 선택: Seiries 반환
display(df['A'])
print("-df['A']-")

# 열 선택: DataFrame 반환
display(df[['A']])
print("-df[['A']]-")

# 행 선택
display(df[0:5])
print("-df[0:5]-")
```

```python
# loc: 열과 행을 이름으로 지정 가능
display(df.loc[0:7, ['A']])
print("-df.loc[0:7, ['A']]-")

# 열 선택: DataFrame 반환
display(df.loc[:, ['A']])
print("-df.loc[:, ['A']]-")

# 행 선택
display(df.loc[3:8])
print("-df.loc[3:8]-")

# 단일 행 선택
display(df.loc[3:3])
print("-df.loc[3:3]-")
```

```python
# iloc: 열과 행을 이름으로 지정 가능
display(df.iloc[0:7, [0]])
print("-df.iloc[0:7, [0]]-")

# 열 선택: DataFrame 반환
display(df.iloc[:, [0]])
print("-df.iloc[:, [0]]-")

# 행 선택
display(df.iloc[3:8])
print("-df.iloc[3:8]-")

# 단일 행 선택
display(df.iloc[3:4])
print("-df.iloc[3:4]-")
```

```python
# 복수 조건을 이용한 데이터 선택
# AND 조건
display(df.query('A>0.5 & B < 0.0'))
print("-df.query('A>0.5 & B < 0.0')-")

# OR 조건
display(df.query('A>0.5 | B<0.0'))
print("-df.query('A>0.5 | B<0.0')-")
```

### 3.
아래와 같은 샘플 데이터 df를 이용해 filter를 실행하라.  
데이터 : 직접 생성한 아래와 같은 df 데이터 (변수 4개, 자료수 4개)

```python
df = pd.DataFrame({'ABC': ['A','B','C','A'], 
	'abc':['a','b','c','c'],
	'group1':['G1','G2','G3','G4'],
	'value1':[11,13,12,14],
	'val1_mod':[0.231,0.197,0.213, 0.199] })
df
```

```python
# 칼럼이름 직접 입력
display(df.filter(items=['group1', 'value1']))
print("-df.filter(items=['group1', 'value1'])-")

# regex(regular express) 이용. '1' 포함
display(df.filter(regex='1', axis=1))
print("-df.filter((regex='1', axis=1)-")

# regex이용. '1'로 끝나는 컬럼
display(df.filter(regex='1$', axis=1))
print("-df.filter(regex='1$', axis=1)-")

# 'a' 포함된 컬럼. regex='a'와 동일
display(df.filter(like='a', axis=1))
print("-df.filter(like='a', axis=1)-")

# 'A' 또는 'a' 포함된 칼럼([] 이용)
display(df.filter(regex='[Aa]', axis=1))
print("-df.filter(regex='[Aa]', axis=1)-")
```

```python
# 조건 입력 및 mask 적용
display(df.mask(df.value1>12))
print("-df.mask(df.value1>12)-")

display(df.mask(df.abc=='a'))
print("-df.mask(df.abc=='a')-")

# 조건 입력 및 mask 적용,
# 조건에 해당되지 않는 값 처리
display(df.mask(cond=df.value1>12, other=999))
print("-df.mask(cond=df.value1>12, other=999)-")
```

### 4.
아래와 같은 샘플 데이터 df1, df2, df3을 이용해 concat(세로 결합)을 실행하라.  
데이터 : 직접 생성한 아래와 같은 df1, df2, df3 데이터 (변수 4개, 자료 수 4개)

```python
# df1 df2, df3 데이터 생성
# df1, df2, df3 데이터 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
		'B': ['B0', 'B1', 'B2', 'B3'],
		'C': ['C0', 'C1', 'C2', 'C3'],
		'D': ['D0', 'D1', 'D2', 'D3']},
		index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
		'B': ['B4', 'B5', 'B6', 'B7'],
		'C': ['C4', 'C5', 'C6', 'C7'],
		'D': ['D4', 'D5', 'D6', 'D7']},
		index=[4, 5, 6, 7])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
		'B': ['B8', 'B9', 'B10', 'B11'],
		'C': ['C8', 'C9', 'C10', 'C11'],
		'D': ['D8', 'D9', 'D10', 'D11']},
		index=[8, 9, 10, 11])

display(df1)
print("-df1-")

display(df2)
print("-df2-")

display(df3)
print("-df3-")
```

```python
# axis = 1: 세로 결합이 아닌 가로 결합
result=pd.concat([df1, df2, df3], axis=1)

# concat 결과 데이터 result 확인
display(result)
print("-result-")

# ignore_index = True: 기존 index가 아닌 새로운 index로 재생성
result=pd.concat([df1, df2, df3], axis=1, ignore_index=True)
display(result)
print("-result(새로운 index 재생성)-")

# concat 결과 데이터의 변수명 생성
result.columns=["col_"+str(i) for i in range(12)]

# concat 결과 데이터 result 확인
display(result)
print("-result(변수명 생성됨)-")
```

### 6.
아래와 같은 샘플 데이터 left, right을 이용해 merge를 실행하라.  
데이터 : 직접 생성한 아래와 같은 left, right 데이터 (변수 4개, 자료수 4개)

```python
# left, right 데이터 생성
left = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
		'key2': ['K0', 'K1', 'K0', 'K1'],
		'A': ['A0', 'A1', 'A2', 'A3'],
		'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K2', 'K3'],
		'key2': ['K0', 'K0', 'K0', 'K0'],
		'C': ['C0', 'C1', 'C2', 'C3'],
		'D': ['D0', 'D1', 'D2', 'D3']})

display(left)
print("-left-")

display(right)
print("-right-")
```

```python
# mergge 결과 result 생성
result=pd.merge(left, right, on=["key1", "key2"])
display(result)
print("-result(inner)-")

# how=left
result=pd.merge(left, right, on=["key1", "key2"], how='left')
display(result)
print("-result(left)-")

# how=right
result=pd.merge(left, right, on=["key1", "key2"], how='right')
display(result)
print("-result(right)-")

# how=outer
result=pd.merge(left, right, on=["key1", "key2"], how='outer')
display(result)
print("-result(outer)-")

# indicator
result=pd.merge(left, right, on=["key1", "key2"], how='outer', indicator='True')
display(result)
print("-result(outer, indicator)-")
```

### 7.
아래와 같은 샘플 데이터 left, right을 이용해 join를 실행하라.  
데이터: 직접 생산한 아래와 같은 left, right 데이터  
(left:변수 3개, 자료수 4개, right: 변수 2개, 자료 수 3개)

```python
# left, right 데이터 생성
left = pd.DataFrame({'key1': [1, 2, 3, 4],
                    	'A': ['A0', 'A1', 'A2', 'A3'],
                   	 'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'C': ['C1', 'C2', 'C5'],
                 		 'D': ['D1', 'D2', 'D5']},
                    	index=[1, 2, 5])


# join 결과 result 생성
result=left.join(right)
display(result)
print("-result(left)-")

# how=right
result=left.join(right, how='right')
display(result)
print("-result(right)-")

# how=inner
result=left.join(right, how='inner')
display(result)
print("-result(inner)-")

# how=outer
result=left.join(right, how='outer')
display(result)
print("-result(outer)-")

# on:join의 key를 지정(left의 column명)
result=left.join(right, on='key1')
display(result)
print("-result(key 지정)-")

# merge로 결합
result2=pd.merge(left, right, left_on='key1', right_index=True, how='left')
display(result2)
print("-result2(merge로 결합)-")
```

### 9. 
아래와 같은 샘플 데이터 df을 이용해 pivot과 melt를 실행하라.  
데이터: 직접 생성한 아래와 같은 df 데이터 (변수 4개, 자료 수 6개)

```python
# df 데이터 생성
df = pd.DataFrame({'index': [1,1,1,2,2,2],
	'column': ['col_1', 'col_2', 'col_3', 'col_1', 'col_2', 'col_3'],
	'value_1': [1, 2, 3, 4, 5, 6],
	'value_2': ['x', 'y', 'z', 'q', 'w', 't']})
display(df)
print("-df-")

# pivot: index, column, values를 지정하여 pivot
display(df.pivot(index='index', columns='column', values='value_1'))
print("-pivot_df-")

# 동일한 기능 다른 표현
display(df.pivot(index='index', columns='column')['value_1'])
print("-pivot_df(다른 코딩 표현)-")

# 복수의 pivot column 지정
display(df.pivot(index='index', columns='column', values=['value_1', 'value_2']))
print("-복수의 column 지정된 pivot_df-")

#pivot column 미지정
display(df.pivot(index='index', columns='column'))
print("-column 미지정된 pivot_df-")

# index 미지정 시 변환되는 데이터 구조 주의
display(df.pivot(columns='column', values='value_1'))
print("-index 미지정된 pivot_df-")

# melt: id_vars를 제외한 나머지 컬럼을 풀어서 쌓음
display(df.melt(id_vars=['index', 'column']))
print("-melt_df-")

# id 및 value 컬럼 지정
display(df.melt(id_vars=['index', 'column'], value_vars=['value_1']))
print("-id, value 지정된 melt_df-")
```
