# 1. 데이터 선택(p.18)
정규분포에서 추출한 10기이 무작위 데이터로 slice, loc, iloc, query 등을 이용하여 데이터를 선택하라.  
데이터: 정규분포에서 추출한 10개의 무작위 데이터.

- (변수: 4개, 자료 수: 10개, random seed=1234)

```python
# 샘플 데이터 통이를 위해 seed 고정: 1234
np.random.seed(1234)

# 랜덤 함수를 이용해 df데이터 생성
df=pd.DataFrame(randn(10,4), columns=['A', 'B', 'C', 'D'])
df
```

```python
# 열 선택: Seiries 반환
display(df['B'])
print("-df['B']-")

# 열 선택: DataFrame 반환
display(df[['B', 'D']])
print("-df[['B', 'D']]-")

# 행 선택
display(df[4:9])
print("-df[4:9]-")
```

```python
# loc: 열과 행을 이름으로 지정 가능
display(df.loc[2:8, ['A', 'C']])
print("-df.loc[2:8, ['A', 'C']]-")

# 열 선택: DataFrame 반환
display(df.loc[:, ['A', 'C']])
print("-df.loc[:, ['A', 'C']]-")

# 행 선택
display(df.loc[3:8])
print("-df.loc[3:8]-")

# 단일 행 선택
display(df.loc[3:3])
print("-df.loc[3:3]-")
```

```python
# iloc: 열과 행을 이름으로 지정 가능
display(df.iloc[0:7, [0]])
print("-df.iloc[0:7, [0]]-")

# 열 선택: DataFrame 반환
display(df.iloc[:, [0]])
print("-df.iloc[:, [0]]-")

# 행 선택
display(df.iloc[2:7])
print("-df.iloc[2:7]-")

# 단일 행 선택
display(df.iloc[5:6])
print("-df.iloc[5:6]-")
```

```python
# 조건을 이용한 데이터 선택
display(df.query('D>0.5'))
print("-df.query('D>0.5')-")

# query와 동일한 slice 표현
display(df[df.B>df.D])
print("-df[df.B>df.D]-")

# 복수 조건을 이용한 데이터 선택
# AND 조건
display(df.query('A>0.5 & D<0.0'))
print("-df.query('A>0.5 & D<0.0')-")

# OR 조건
display(df.query('A>0.5 | D<0.0'))
print("-df.query('A>0.5 | D<0.0')-")
```

# 2. 데이터 결합(p.32)
아래와 같은 샘플 데이터 df4, df5, df6을 이용해 concat(좌우 결합)을 실행하라.  
데이터 : 직접 생성한 아래와 같은 df1, df2, df3 데이터 (변수 4개, 자료 수 4개)

```python
# df4 df5, df6 데이터 생성
# df4, df5, df6 데이터 생성
df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
		'B': ['B0', 'B1', 'B2', 'B3'],
		'C': ['C0', 'C1', 'C2', 'C3'],
		'D': ['D0', 'D1', 'D2', 'D3']},
		index=[0, 1, 2, 3])
df5 = pd.DataFrame({'E': ['E4', 'E5', 'E6', 'E7'],
		'F': ['F4', 'F5', 'F6', 'F7'],
		'G': ['G4', 'G5', 'G6', 'G7'],
		'H': ['H4', 'H5', 'H6', 'H7']},
		index=[0, 1, 2, 3])
df6 = pd.DataFrame({'I': ['I8', 'I9', 'I10', 'I11'],
		'J': ['J8', 'J9', 'J10', 'J11'],
		'K': ['K8', 'K9', 'K10', 'K11'],
		'L': ['L8', 'L9', 'L10', 'L11']},
		index=[0, 1, 2, 3])

display(df4)
print("-df4-")

display(df5)
print("-df5-")

display(df6)
print("-df6-")
```

```python
# axis = 1: 세로 결합이 아닌 가로 결합
result=pd.concat([df4, df5, df6], axis=1)

# concat 결과 데이터 result 확인
display(result)
print("-result-")

# ignore_index = True: 기존 index가 아닌 새로운 index로 재생성
result=pd.concat([df4, df5, df6], axis=1, ignore_index=True)
display(result)
print("-result(새로운 index 재생성)-")

# concat 결과 데이터의 변수명 생성
result.columns=["col_"+str(i) for i in range(12)]

# concat 결과 데이터 result 확인
display(result)
print("-result(변수명 생성됨)-")
```

# 4. 데이터 변환(p.54)
아래는 운동에 따른 혈당 산소 요구량 데이터이다.  
pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 아래와 같이 변환하라.
- 데이터: FITNESS.CSV(변수: 10개, 자료 수: 31개)

```python
fit3=pd.read_csv("../data/FITNESS.CSV", engine="python")
fit4=fit3[["GENDER","AGEGROUP","WEIGHT"]]
fit4.head()
```

```python
df_group = fit4.groupby(['GENDER', 'AGEGROUP']).agg('mean')
df_group.reset_index(inplace = True)
display(df_group)
print('-df_group(agg 함수 사용)-')

df_group2=fit4.groupby(["GENDER","AGEGROUP"]).mean()
df_group2.reset_index(inplace=True)
display(df_group2)
print("-df_group2-")
```

```python
df_group3=df_group.pivot(index="GENDER",columns="AGEGROUP",values="WEIGHT")
df_group3
```

```python
df_group4=df_group.pivot(index="AGEGROUP",columns="GENDER",values="WEIGHT")
df_group4
```

# 5. 데이터 조작(p.65)
서울시 구별 인구수와 CCTV 설치 비율에 대한 분석을 이해 공공데이터를 확인하고 적절한 방법으로 처리하시오.  
최근 3년간 cctv 증가율과 한국인, 외국인, 고령자의 비율에 따른 데이터 변화를 보고자 한다

```python
# 탭으로 분리된(tsv) .txt 텍스트파일 불러오기
# data = pd.read_csv('파일경로', sep = "\t", , engine='python', encoding = "인코딩방식")
pop_seoul = pd.read_csv('../data/서울인구통계.txt',sep="\t", engine='python', thousands = ',',encoding = "utf-8")
pop_seoul.head()
```

```python
# 자치구별 인구합계, 한국인 합계, 외국인 합계, 고령자 데이터만을 추출
# index 0, 1행은 중복으로 필요 없는 행으로 삭제
pop1=pop_seoul[['자치구','인구','인구.3','인구.6','65세이상고령자']]
pop_seoul=pop1.drop([0,1])
pop_seoul.head()
```

```python
# 컬럼명을 구분할 수 있도록 리네임
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                         pop_seoul.columns[1]:'인구수',
                         pop_seoul.columns[2]:'한국인',
                         pop_seoul.columns[3]:'외국인',
                         pop_seoul.columns[4]:'고령자'},inplace=True)
pop_seoul.head()
```

```python
# 서울시 CCTV 엑셀파일을 열고 5번째까지 자료를 읽기
cctv_seoul=pd.read_excel('../data/서울시cctv.xlsx')
cctv_seoul.head()
```

```python
# 칼럼의 데이터를 확인
cctv_seoul.columns
```

```python
# 인구통계 자료와 통일을 하기 위해 기관명을 구별로 인덱스 이름 변경
cctv_seoul.rename(columns={cctv_seoul.columns[0] : '구별'},inplace=True)
# inplace=TRUE로 해야지 값까지 다 포함해서 가져감
cctv_seoul.rename(columns={cctv_seoul.columns[1] : 'cctv수'},inplace=True) 
# 소계를 cctv수로 이름 변경
cctv_seoul.head()
```

```python
# sort_values : 값을 기준으로 정렬, by=기준 열
cctv_seoul.sort_values(by='cctv수',ascending=True).head(10)  #True 오름차순
```

```python
cctv_seoul.sort_values(by='cctv수',ascending=False).head(10) #False 내림차순
```

```python
# NaN 값 확인
cctv_seoul.isnull().sum() #NaN 값이 있는 개수를 확인 가능
```

```python
cctv_seoul.loc[cctv_seoul.isnull()['2011년 이전']] # 2011년 이전 열을 기준으로 NaN 값 추출
```

```python
# 최근 3년간 cctv 증가율 비교
# 증가율 = 2016~2018 cctv 수/2016년 이전 cctv 수
cctv_seoul['최근증가율']=(cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년'])/(cctv_seoul['cctv수']-(cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년']))*100
cctv_seoul.sort_values(by='최근증가율',ascending=False).head(5)
```

```python
# 인구 데이터 확인
pop_seoul.head()
```

```python
# index 2의 행 삭제를 위해 [2] 입력
pop_seoul.drop([2],inplace=True)
pop_seoul.head()
```

```python
# 데이터 타입 확인.
pop_seoul.dtypes
```

```python
# object로 연산이 불가 하기 때문에 int, float형으로 변경 필요
pop_seoul=pop_seoul.astype({"인구수":"int","한국인":"int","외국인":"int","고령자":"int"})
pop_seoul.dtypes
```

```python
# 구별 ‘외국인비율＇과 ‘고령자비율＇을 계산
pop_seoul['외국인 비율']=pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자 비율']=pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head()
```

```python
# 전체 ‘외국인비율‘,’외국인＇로 정렬
pop_seoul.sort_values(by='외국인 비율', ascending=False).head()
```

```python
pop_seoul.sort_values(by='고령자 비율', ascending=False).head()
```

```python
# cctv_seoul의 “구별” 열의 내용이 글자사이 공백으로 인해 병합이 되지 않기 때문에 rename 실시
cctv_seoul['구별']=['강남구','강동구','강북구','강서구','관악구','광진구','구로구','금천구','노원구','도봉구',
'동대문구','동작구','마포구','서대문구','서초구','성동구','성북구','송파구','양천구','영등포구','용산구','은평구',
'종로구','중구','중랑구']
```

```python
# merge를 사용하여 병합
result = pd.merge(cctv_seoul, pop_seoul, on='구별')
result
```

```python
# 의미없는 컬럼을 del 명령어로 제거
# 일반적으로 행을 삭제할때는 drop, 열을 삭제할때는 del을 사용
# 2011 ~ 18년까지의 cctv수는 의미가 없으므로 삭제

del result['2011년 이전']
del result['2012년']
del result['2013년']
del result['2014년']
del result['2015년']
del result['2016년']
del result['2017년']
del result['2018년']
result.head()
```
